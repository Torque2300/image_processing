```python
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
sns.set(style="white", color_codes=True)
```

```python
class KNN:
    def __init__(self, k):        
        self.k = k
    
    def fit(self, x, y):
        self.train_X = x
        self.train_y = y
        
    def predict(self, x):
        dists = self.compute_distances(x)        
        pred = np.zeros(x.shape[0])
        for i in range(x.shape[0]):
            indices = dists[i].argsort()[:self.k]
            pred[i] = np.argmax(np.bincount(self.train_y[indices]))
        return pred
        
    def compute_distances(self, x):
        num_test = x.shape[0]
        num_train = self.train_X.shape[0]
        dists = np.zeros((num_test, num_train))
        dists = np.sum(np.abs(x[:,None,:] - self.train_X[None,:,:]),axis=2)
        return dists
```

```python
from sklearn.datasets import load_digits
mnist = load_digits()
print(mnist.data.shape)
X = mnist.data 
y = mnist.target
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=123)
print(X_train.shape, y_train.shape)
print(X_test.shape, y_test.shape)
```

```python
from sklearn.metrics import accuracy_score


k = np.arange(1, 50, 1)
accuracy = []
for i in range(len(k)):
    model = KNN(k[i])
    model.fit(X_train, y_train)
    pred = model.predict(X_test)
    true = np.sum([1 if pred[i] == y_test[i] else 0 for i in range(len(pred))])
    accuracy.append(true/len(pred))
```

```python
print('K =', np.argmax(accuracy), '\naccuracy =', accuracy[np.argmax(accuracy)])
```

```python
model = KNN(2)
model.fit(X_train, y_train)
pred = model.predict(X_test)
```
