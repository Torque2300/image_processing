```python
import numpy as np
import cv2 as cv
from matplotlib import pyplot as plt
from ipywidgets import interact
import ipywidgets as widgets
```

```python
image = cv.imread('aaa.jpg')
gray = cv.cvtColor(image, cv.COLOR_BGR2GRAY)
is_plot = True
@widgets.interact_manual(
    lt=(0, 255), ht=(0, 255), k=(1,20))
def biggest_object_area(lt=100, ht=100, k=7):
    edged = cv.Canny(gray.copy(), lt, ht) 
    kernel = cv.getStructuringElement(cv.MORPH_RECT, (k, k))
    closed = cv.morphologyEx(edged, cv.MORPH_CLOSE, kernel)
    contours, hierarchy = cv.findContours(closed.copy(), cv.RETR_EXTERNAL, cv.CHAIN_APPROX_SIMPLE)
    sorted_contour = sorted(contours, key=lambda x: cv.contourArea(x), reverse=True)
    largestcontour = sorted_contour[0]
    s = cv.contourArea(largestcontour) / (image.shape[0] * image.shape[1])
    results = image.copy()
    cv.drawContours(results, [largestcontour],0,(0,0,255),5)
    if (is_plot):
        plt.imshow(results, cmap='gray'), plt.xticks([]), plt.yticks([])
        plt.show()
    return largestcontour
```

```python
def biggest(img, lt, ht, k, is_plot=False):
    edged = cv.Canny(img.copy(), lt, ht) 
    kernel = cv.getStructuringElement(cv.MORPH_RECT, (k, k))
    closed = cv.morphologyEx(edged, cv.MORPH_CLOSE, kernel)
    contours, hierarchy = cv.findContours(closed.copy(), cv.RETR_EXTERNAL, cv.CHAIN_APPROX_SIMPLE)
    sorted_contour = sorted(contours, key=lambda x: cv.contourArea(x), reverse=True)
    largestcontour = sorted_contour[0]
    s = cv.contourArea(largestcontour) / (image.shape[0] * image.shape[1])
    results = img.copy()
    cv.drawContours(results, [largestcontour],0,(0,0,255),5)
    if (is_plot):
        plt.imshow(results, cmap='gray'), plt.xticks([]), plt.yticks([])
        plt.show()
    return largestcontour, results
```

```python
image = cv.imread('chipmunk\\4.jfif')
gray = cv.cvtColor(image, cv.COLOR_BGR2GRAY)
fbur, fi = biggest(gray, 132, 242, 3)
# plt.imshow(fi)

image2 = cv.imread('chipmunk\\6.jpg')
gray2 = cv.cvtColor(image2, cv.COLOR_BGR2GRAY)
sbur, si = biggest(gray2, 100, 116, 7)
# plt.imshow(si)

image = cv.imread('badger\\22.png')
gray = cv.cvtColor(image, cv.COLOR_BGR2GRAY)
fbad, ti = biggest(gray, 55, 91, 5)
# plt.imshow(ti)

image = cv.imread('badger\\33.jfif')
gray = cv.cvtColor(image, cv.COLOR_BGR2GRAY)
sbad, fti = biggest(gray, 55, 91, 5)

bur = cv.matchShapes(fbur, sbad,1,0.0)
fc = cv.matchShapes(fbad, sbad,1,0.0)

print(bur, fc)
if (bur > fc):
    print("barsuk")
else:
    print("burunduk")

plt.imshow(fti), plt.xticks([]), plt.yticks([])
plt.show()
```
